```python
import pandas as pd
import numpy as np
```

```python
file = pd.read_csv("/home/bunny/Downloads/dataset_2191_sleep.csv")
file
```

```python
file.dtypes
```

```python
file['max_life_span'] = file['max_life_span'].replace('?', np.nan)
file['gestation_time'] = file['gestation_time'].replace('?',np.nan)
file['total_sleep'] = file['total_sleep'].replace('?',np.nan)
```

```python
file = file.astype(float)
```

```python
file.dtypes
```

```python
filter = ['body_weight','brain_weight','max_life_span','gestation_time','predation_index','sleep_exposure_index','total_sleep']
x = file[filter]
y = file['danger_index']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
trainx,valx,trainy,valy = train_test_split(x,y,random_state=1)
trainx = trainx.replace(np.nan, 2.71)
valx = valx.replace(np.nan, 2.71)
trainy = trainy.replace(np.nan, 2.71)
valy = valy.replace(np.nan, 2.71)
```

```python
model = GaussianNB()
model.fit(trainx, trainy)
```

```python

print(valy, model.predict(valx))
```
